```python
#  2 * x + 5 = 20
import numpy as np
def equation_func(solution):
  x = solution[0]
  equation_output = 2 * x + 5
  target_output = 20
  fitness = 1.0 / np.abs(equation_output - target_output  + 1e-8)
  return fitness
```

```python
lista = [-10,-5,0,5,7.5, 10, 11]
for value in lista:
  result = equation_func([value])
  print(result)
```
