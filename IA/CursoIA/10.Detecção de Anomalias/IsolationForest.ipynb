```python
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.datasets import load_iris
```

```python
data = load_iris().data
```

```python
model = IsolationForest(contamination=0.01)
model.fit(data)
```

```python
predictions = model.predict(data)
```
